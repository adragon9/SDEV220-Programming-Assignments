# 3. Update

### 3.1 Dropping Rows

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

### 3.2 Dropping Columns

```python
df.drop('Header 3', axis=1)
```

### 3.3 Creating Calculated Columns

```python
df['New Col'] = df['Header 1'] + df['Header 2']
```

```python
df.head()
```

### 3.4 Updating an Entire Column

```python
df['New Col'] = 100
```

### 3.5 Updating a Single Value

```python
df.iloc[0, -1] = 10
```

```python
df['Header 1 Rev'] = df['Header 1'].apply(lambda x: 6 if x==1 else random.randint(0, 100))
```
